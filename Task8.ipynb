### **CORRELATION MATRIX ***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'train' with your dataset
train = pd.read_csv('/content/segmentation data.csv')

# Calculate the correlation matrix
correlation_matrix = train.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix - Entire Train Dataset')
plt.show()
```

```python
N = 10

# Calculate absolute correlation values for each feature
correlation_values = correlation_matrix.abs()

# Sort the correlation values in descending order
sorted_correlation = correlation_values.unstack().sort_values(ascending=False)

# Exclude self-correlations and duplicate correlations
sorted_correlation = sorted_correlation[sorted_correlation.index.get_level_values(0) != sorted_correlation.index.get_level_values(1)]

# Select the top 10 features with highest absolute correlation values
top_features = sorted_correlation[:N]

print("Top", N, "features based on correlation:")
print(top_features)
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("/content/segmentation data.csv")

# Select features
features = ["Age", "Income"]  # Replace with your desired features
data_to_cluster = data[features]

# Scale features (optional, but recommended for DBSCAN)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_to_cluster)

# Define parameters
eps = 0.5  # Adjust based on your analysis
min_samples = 5  # Adjust based on your analysis

# Perform DBSCAN clustering
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
clusters = dbscan.fit_predict(scaled_data)

# Plot the clusters
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clusters)
plt.title("DBSCAN Clusters")
plt.show()

# Create a dictionary to store cluster information
cluster_info = {}
for i in range(len(clusters)):
   cluster_label = clusters[i]
   if cluster_label not in cluster_info:
       cluster_info[cluster_label] = {"size": 0, "data_points": []}
   cluster_info[cluster_label]["size"] += 1
   cluster_info[cluster_label]["data_points"].append(i)

# Print cluster information
for cluster_label, info in cluster_info.items():
   size = info["size"]
   data_points = info["data_points"]
   print(f"Cluster {cluster_label}:")
   print(f"- Size: {size}")
   print(f"- Data points: {data_points}")


```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Load data
data = pd.read_csv("/content/segmentation data.csv")

# Select features
features = ["Age", "Income"]  # Replace with your desired features
data_to_cluster = data[features]

# Scale features (optional, but recommended for DBSCAN)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_to_cluster)

# Define parameter ranges
eps_values = [0.1,0.5,1.0]
min_samples_values = [3, 5, 7,]

# Create empty lists to store results
cluster_labels, silhouette_scores = [], []

# Iterate over parameter combinations
for eps in eps_values:
    for min_samples in min_samples_values:
        # Perform DBSCAN clustering
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        clusters = dbscan.fit_predict(scaled_data)

        # Store cluster labels
        cluster_labels.append(clusters)

        # Calculate silhouette score
        silhouette_score_val = silhouette_score(scaled_data, clusters)
        silhouette_scores.append(silhouette_score_val)

        # Print silhouette score for each combination
        print(f"eps: {eps}, min_samples: {min_samples}, Silhouette Score: {silhouette_score_val}")

        # Visualize clusters (optional)
        plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clusters)
        plt.title(f"DBSCAN - eps: {eps}, min_samples: {min_samples}")
        plt.show()
  # Find best parameters based on silhouette score
best_index = np.argmax(silhouette_scores)
best_eps = eps_values[best_index // len(min_samples_values)]
best_min_samples = min_samples_values[best_index % len(min_samples_values)]

print(f"Best parameters: eps = {best_eps}, min_samples = {best_min_samples}")

# Perform final clustering using best parameters
best_dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
final_clusters = best_dbscan.fit_predict(scaled_data)




# Perform final clustering using best parameters (optional)
# Uncomment the following lines if you want to perform final clustering
best_dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
final_clusters = best_dbscan.fit_predict(scaled_data)

# Use final clusters for further analysis...
```
