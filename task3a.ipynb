# Trimming
This is the first way we can remove outliers in correct meaning as we don’t give them other certain values or treat them in another way. Trimming removes all the bad data from the dataset just to ensure the quantity of data is not much; otherwise, for analysis, we won’t have much data for ML model development.

```python
new_df = df[df['age'] < upper_limit]
new_df.shape
```

```python
#COMPARING
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(df['age'], kde=True)

plt.subplot(2,2,2)
sns.boxplot(data = df['age'])

plt.subplot(2,2,3)
sns.histplot(new_df['age'], kde=True)

plt.subplot(2,2,4)
sns.boxplot(data = new_df['age'])

plt.show()
```

# **CAPPING**
Capping is a second way to impute the outliers with some other values. There can be mean, median or mode or any constant value also (that we gonna do here) leads to the condition where there will be no outliers in the dataset

```python
new_df_cap = df.copy()

new_df_cap['age'] = np.where(
    new_df_cap['age'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['age'] < lower_limit,
        lower_limit,
        new_df_cap['age']
    )
)
```

```python
new_df_cap.shape
```

```python
#COMPARING

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(df['age'], kde=True)

plt.subplot(2,2,2)
sns.boxplot(data = df['age'])

plt.subplot(2,2,3)
sns.histplot(new_df_cap['age'], kde=True)

plt.subplot(2,2,4)
sns.boxplot(data = new_df_cap['age'])

plt.show()
```
