# **PLOTTING CAR DATA**

*   Telsa, Model Y have highest popularity in the car model.




```python
Car_Popularity = data.value_counts(["Make","Model"],ascending=True)
Car_Popularity.tail(10).plot(color='red',kind='barh',title = 'Car Model Popularity Graph',figsize=(10,5), ylabel = "TOP 10 POPULAR MODELS", xlabel = "Total number of cars", fontsize = 15)
```

```python
CarsPerCounty = data.value_counts(["County"],ascending=True)
CarsPerCounty.tail(10).plot(kind='barh', title = 'Cars Per County Graph',figsize=(10,5),fontsize = 15, ylabel = "County", xlabel = "Total number of cars")
```

```python
CarsPerCity = data.value_counts(["City"],ascending = True)
CarsPerCity.tail(10).plot(color='yellow',kind='barh', title = 'Cars Per City Graph',figsize=(10,5),fontsize = 15, ylabel = "City", xlabel = "Total number of cars")
```

```python
CarsPerCityandCounty = data.value_counts(["City","County"],ascending = True)
CarsPerCityandCounty.tail(10).plot(color='green',kind='barh', title = 'Cars Per City and County Graph',figsize=(10,5),fontsize = 15, ylabel = "City and County", xlabel = "Total number of cars")
```

```python
import numpy as np
carpopularity = data.value_counts(["Make","Model"],ascending=True)
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
models = ['TESLA MODEL 3','TESLA MODEL Y','tesla moddel x','tesla model s']
counting = [ 16237,18066,3265,4139]
ax.bar(models,counting)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.array([1,1,1,1,1, 3265, 4139, 6457,16237, 18066])
plt.hist(a)
ax.set_xlabel('popularity')
ax.set_ylabel('cars')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
cars=['tesla model 3','tesla model y','tesla moddel x','tesla model s']
popularity=[16237,18066,3265,4139]
ax.pie(popularity,labels=cars)
plt.show()
```
