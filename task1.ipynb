# **LOGIC**

#### Here, the no.of customers leaving the bank i.e 'churn' should be predicted. For this prediction let say if the customer is not active then it will raise the case of 'churn'. From the data set, if the 'active_member' is '1' then there is no case of churn i.e 'churn' will be '0' and viceversa. In other cases, 'churn' may also depend on 'tenure', 'balance' etc.

```python
import pandas as pd
df['churn'] = 1
df.loc[df['active_member'] == 1, 'churn'] = 0
print(df.head())
```

**Changing the columns by considering if the member is active then there is no case of 'churn'**
